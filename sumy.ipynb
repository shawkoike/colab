```python
from __future__ import absolute_import
from __future__ import division, print_function, unicode_literals

from sumy.parsers.html import HtmlParser
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer as Summarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words


LANGUAGE = "japanese"
SENTENCES_COUNT = 5

document = '''
  初めまして、中嶋祥汰と申します！
愛知在住の24歳です。

医療系の専門学校を卒業後、東京の大手整骨院へ柔道整復師として就職。その後、人間関係のストレスから逃れるように独立し、美容整体院を6店舗立ち上げました。

しかし、集客が上手くいかず倒産。
その経験もあってWeb集客の大切さを学び、現在は自身の強みを活かして、医療・美容業界を中心に人材教育、Webコンサルタントとしてご支援させて頂いております。

また営業マン7名・クリエイター3名が在籍しているフリーランス集団INCOMEの代表もやっております。

主な実績は下記で紹介しておきます。
※長くなりますので割愛あり

【主な実績】
・大手広告企業が運営する月間2000万PVのメディアで専門家として活動
・治療院やパーソナルジムのLP制作・SEOコンサル・広告運用のご支援
・治療院のMEOコンサルティングを担当。3ヶ月後集客率18倍、過去最高利益達成。
・中小企業様のリスティング広告運用のご支援（運用額50万〜200万）
・記事LPの制作・Facebook広告の運用（CV6件から12件へUP）
・Webライターとして文字単価0.3円から1文字6円までUP
'''

parser = PlaintextParser.from_string(document, Tokenizer(LANGUAGE))
stemmer = Stemmer(LANGUAGE)

summarizer = Summarizer(stemmer)
summarizer.stop_words = get_stop_words(LANGUAGE)

for sentence in summarizer(parser.document, SENTENCES_COUNT):
  print(sentence)
```
